```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
phi0 = 1
a = 1
```

```python
def psi(mu):
    if -1 <= mu <= 0:
        return phi0 + a*mu**4
    elif 0 <= mu <= 1:
        return phi0
    
def psi_p1(mu):
    return 0.5*(2*phi0 - (a*mu/4) + (a/5))

def psi_p3(mu):
    return psi_p1(mu) + 5*a/8*(3/7 - 1/5)*(3*mu**2 - 1) + 7*a/8*(1/2 - 5/8)*(5*mu**3 - 3*mu)
```

```python
m = np.linspace(-1,1,num=100)
```

```python
psi_m = [psi(mu) for mu in m]
```

```python
plt.plot(m, psi_m, label=r'$\psi(\mu)$')
plt.plot(m, psi_p1(m), label=r'$\psi_{P_1}(\mu)$')
plt.plot(m, psi_p3(m), label=r'$\psi_{P_3}(\mu)$')
plt.legend()
plt.xlabel(r'$\mu$')
```
